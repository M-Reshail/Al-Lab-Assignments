```python
import random

def choose_word():
    words = ['python', 'programming', 'hangman', 'computer', 'challenge', 'openai']
    return random.choice(words)

def display_word(word, guessed_letters):
    return ' '.join([letter if letter in guessed_letters else '_' for letter in word])

def print_hangman(attempts_left):
    hangman_stages = [
        """
           -----
           |   |
           O   |
          /|\\  |
          / \\  |
               |
        _______|_
        """,
        """
           -----
           |   |
           O   |
          /|\\  |
          /    |
               |
        _______|_
        """,
        """
           -----
           |   |
           O   |
          /|\\  |
               |
               |
        _______|_
        """,
        """
           -----
           |   |
           O   |
          /|   |
               |
               |
        _______|_
        """,
        """
           -----
           |   |
           O   |
           |   |
               |
               |
        _______|_
        """,
        """
           -----
           |   |
           O   |
               |
               |
               |
        _______|_
        """,
        """
           -----
           |   |
               |
               |
               |
               |
        _______|_
        """
    ]
    print(hangman_stages[6 - attempts_left])

def hangman():
    word_to_guess = choose_word()
    guessed_letters = set()
    attempts = 6
    print("Welcome to Hangman!")
    
    while attempts > 0:
        print_hangman(attempts)
        print(f"\nWord to guess: {display_word(word_to_guess, guessed_letters)}")
        print(f"Guessed letters: {', '.join(sorted(guessed_letters))}")
        print(f"Attempts remaining: {attempts}")
        
        guess = input("Guess a letter: ").lower()

        if len(guess) != 1 or not guess.isalpha():
            print("Please guess a single letter.\n")
            continue

        if guess in guessed_letters:
            print("You already guessed that letter.\n")
        elif guess in word_to_guess:
            print(f"Good job! '{guess}' is in the word.\n")
            guessed_letters.add(guess)
        else:
            print(f"Sorry, '{guess}' is not in the word.\n")
            guessed_letters.add(guess)
            attempts -= 1

        if all(letter in guessed_letters for letter in word_to_guess):
            print_hangman(attempts)
            print(f"\nCongratulations! You've guessed the word '{word_to_guess}' correctly!")
            break
    else:
        print_hangman(attempts)
        print(f"\nGame over! The word was '{word_to_guess}'.")

hangman()
```
