```python
moviesDetails = [
    ("Eternal Sunshine of the Spotless Mind", 200),
    ("Memento", 90),
    ("Requiem for a Dream", 4500),
    ("Pirates of the Caribbean: On Stranger Tides", 3790),
    ("Avengers: Age of Ultron", 365),
    ("Avengers: Endgame", 3560),
    ("Incredibles 2", 20)
]

class Movies:
    def _init_(self, movie_name, budget):
        self.movie_name = movie_name
        self.budget = budget

    def append_data_to_list(self):
        data = (self.movie_name, self.budget)
        moviesDetails.append(data)
        print(moviesDetails)

    def average_of_movies(self):
        if not moviesDetails:
            print("No movies to find budget.")
            return
        
        total_cost = sum(movie[1] for movie in moviesDetails)
        total_movies = len(moviesDetails)
        average_budget = total_cost / total_movies
        print(f"Total Average Budget of your Movies is: {average_budget}")

    def over_budget(self):
        if len(moviesDetails) == 0:
            print("No movies to find.")
            return
        
        limit_of_budget = int(input("Enter the limit of your budget: "))
        total_out_of_budget = 0

        for movie_name, cost_of_movie in moviesDetails:
            if cost_of_movie > limit_of_budget:
                total_out_of_budget += 1
                print(f"{movie_name} is out of budget and the higher budget price is: {cost_of_movie - limit_of_budget}")

        print(f"Total: {total_out_of_budget} movies are out of budget.")

while True:
    choice = input("Press -1 to add movie details\nPress -2 to show over-budget movies\nPress -3 to show average budget\n")
    
    if choice == "1":
        movie = input("Enter Movie Name: ")
        budget = int(input("Enter Movie Budget: "))
        obj1 = Movies(movie, budget)
        obj1.append_data_to_list()

    elif choice == "2":
        obj3 = Movies("", 0)
        obj3.over_budget()

    elif choice == "3":
        obj2 = Movies("", 0)
        obj2.average_of_movies()

    else:
        print("Invalid choice. Please try again.")
```
